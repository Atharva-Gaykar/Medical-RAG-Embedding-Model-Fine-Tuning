```python
import os
from langchain_groq import ChatGroq
import json
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains.llm import LLMChain

import pandas as pd
```

```python


# Ensure your key is set
if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = " "


llma_base_model = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=0.2
)
```

```python
gpt_base_model = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=0.2
)
```

```python


CHUNKS_DIR = r"C:\Users\ATHARVA\Downloads\my codes\python\machine_learning\project_web\Tumor_chunks\pituitary_chunks.json"
START_INDEX = 0
END_INDEX =90

OUTPUT_FILE = "pituitary_qa_dataset.json"
llm = llma_base_model
```

```python

prompt = PromptTemplate(
    input_variables=["page_content"],
    template="""
You are a medical AI assistant.

Given the following medical text, generate high-quality question and answer pairs.
Igonore any non-medical information.
IMP:Igonre if the data only contains questions without answers **Do not** generate questions from such data.

Rules:
- Use ONLY the provided content
- Ignore sentences that do not contain meaningful medical information
- Do NOT hallucinate
- If no useful information exists, return an empty list
- Output MUST be valid JSON only

Output format:
[
  {{
    "question": "...",
    "answer": "..."
  }}
]

Medical Text:
{page_content}
"""
)

qa_chain = LLMChain(
    llm=gpt_base_model,
    prompt=prompt
)
```

```python

all_chunks = []


with open(CHUNKS_DIR, "r", encoding="utf-8") as f:
    data = json.load(f)
    all_chunks.extend(data)
```

```python
from pprint import pprint
pprint(all_chunks[0:2])
```

```python
final_dataset = []
```

```python



# Sort by chunk_id
all_chunks = sorted(all_chunks, key=lambda x: x["metadata"]["chunk_id"])

final_dataset = []

for chunk in all_chunks:
    chunk_id = chunk["metadata"]["chunk_id"]

    if chunk_id < START_INDEX or chunk_id > END_INDEX:
        continue

    page_content = chunk["page_content"].strip()

    if not page_content:
        continue

    try:
        response = qa_chain.invoke(input=page_content)

        qa_pairs = json.loads(response["text"])

        for qa in qa_pairs:
            final_dataset.append({
                "question": qa["question"],
                "answer": qa["answer"],
                "tumor_type": chunk["metadata"]["tumor_type"],
                "chunk_id": chunk_id
            })

    except Exception as e:
        print(f"❌ Error at chunk {chunk_id}: {e}")


```

```python
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(final_dataset, f, indent=2, ensure_ascii=False)

print(f"Dataset created with {len(final_dataset)} Q&A pairs")
print(f"Saved to: {OUTPUT_FILE}")
```
